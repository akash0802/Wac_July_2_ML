## Data preprocessing

```
train.drop('PassengerId', axis = 1, inplace = True)
test.drop('PassengerId', axis = 1, inplace = True)
```

```
def name_preprocess(col, name_title_list):
    for i in col:
        # print(i)
        temp = i.split()
        for j in temp:
            if '.' in j:
                name_title_list.append(j[:-1])
                break
```

```
name_title_train = []
name_title_test = []
name_preprocess(train['Name'], name_title_train)
name_preprocess(test['Name'], name_title_test)
```

```
# name_title_test
```

```
train["Name_title"] = name_title_train
test["Name_title"] = name_title_test
```

```
train['Cabin'].fillna(-999, axis = 0, inplace = True)
test['Cabin'].fillna(-999, axis = 0, inplace = True)
```

```
def cabin_preprocess(col, list_cabin):
    for i in col:
        if i == -999:
            list_cabin.append(i)
        else:
            list_cabin.append(i[0])
```

```
train_cabin = []
test_cabin = []

cabin_preprocess(train['Cabin'], train_cabin)
cabin_preprocess(test['Cabin'], test_cabin)
```

### XGboost

```
# hyper parameter tuning
from sklearn.model_selection import GridSearchCV
```

```
param = {
    'n_estimators' : [100, 500, 700, 1000],
    'min_child_weight' : [1, 2, 3, 4],
    'max_depth' : [2,3,4,5]
}
```

```
grid = GridSearchCV(clf2, param_grid= param, cv = 5)
```

```
grid.fit(X_train, y_train)
```

```
grid.best_score_
```

```
grid.best_params_
```

```
pred = grid.predict(X_val)
```
