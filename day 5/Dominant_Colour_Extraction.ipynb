# Start

```
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```
img = cv2.imread('/content/elephant.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```
img.shape
```

```
plt.imshow(img)
plt.axis('off')
plt.show()
```

```
img.shape
```

```
# making image into a linear array
img_flatten = img.reshape(-1, 3) # 330*500, 3
img_flatten.shape
```

```
from sklearn.cluster import KMeans
```

```
km = KMeans(n_clusters= 10, n_jobs= -1)
```

```
km.fit(img_flatten)
```

```
center = km.cluster_centers_
center # pixel values of centre
```

```
center = np.array(center, dtype= 'int') # to convert floating values into int values
center
```

```
km.labels_
```

```
new_img = np.zeros(img_flatten.shape, dtype= 'uint')
new_img.shape
```

```
plt.imshow(new_img.reshape(330, 500, 3))
plt.axis('off')
plt.show()
```

```
km.labels_[1000]
```

```
center[km.labels_[1000]]
```

```
new_img.shape[0] 
```

```
for i in range(new_img.shape[0]):
    new_img[i] = center[km.labels_[i]]
```

```
plt.imshow(new_img.reshape(330, 500, 3))
plt.axis('off')
plt.show()
```
